```python
from STvAR_Flash import *
from STvAR_FlashX_FiniteDifference import*
from STvAR_FlashX_Advection import *
import numpy as np
```

```python
phi = stvar('PHI',  gridvar = True, vartype = 'VAR', varprefix = 'U')
alpha = stvar('alpha', gridvar = True, vartype = 'RHS', varprefix = 'U')
beta_U = stvarrank1('beta_U', gridvar = True, vartype = 'VAR', varprefix = 'U')

Vec_L = stvarrank1('Vec_L', gridvar = True, vartype = 'RHS', varprefix = 'U')
Mat_LL = stvarrank2('Mat_LL', sym = 'sym01', gridvar = True, vartype = 'VAR', varprefix = 'U')
phi.expr = alpha.isymb**2
```

```python
print(beta_U.symb2isymb())
```

```python
dDDVec_LLL = Vec_L.diff('dDD', Accuracy = 2)
dDDMat_LLLL = Mat_LL.diff('dDD', Accuracy = 2)
```

```python
print(dDDMat_LLLL.symb2isymb().replace('\n', ' & \n').replace('@', ''))
```

```python
dDphi_L = phi.diff('dDD', Accuracy = 4)
dDDVec_LL = Vec_L.diff('dDD')
```

```python
print(dDphi_L.symb2expr().replace('\n     @', ' &\n     '))
```

```python
print(dDDVec_LL.symb2expr().replace('\n     @', ' &\n     '))
```

```python
print(dDDVec_LL.symb2expr())
```
